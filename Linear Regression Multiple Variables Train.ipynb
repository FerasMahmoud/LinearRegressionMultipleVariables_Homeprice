```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:\\Users\\feras\\JupyterNoteBook\\LinearRegressionMultipleVariables\\homeprices.csv')
df
```

```python
df[['bedrooms']] = df[['bedrooms']].fillna(df[['bedrooms']].median()) # fill the missing with median number
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area','bedrooms','age']],df[['price']])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000,3,40]]) # for EX1
```

```python
reg.predict([[2500,4,5]]) # for EX2
```

As you can see less age more cost and more area more cost and more bedrooms more cost
